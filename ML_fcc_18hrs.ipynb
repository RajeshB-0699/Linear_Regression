#y=mx+c; x--> age , y-> charges, m&c ---> weight and bias here y=wx+b

```python
def estimate_charges(age,w,b):
  return w * age + b
```

```python
w=50
b=100
ages = non_smoker_df.age
estimated_charges = estimate_charges(ages,w,b)
```

```python
def try_parameters(w,b):
  ages = non_smoker_df.age
  target = non_smoker_df.charges

  estimated_charges = estimate_charges(ages,w,b)

  plt.plot(ages,estimated_charges,'r',alpha=0.9)
  plt.scatter(ages,target,s=8,alpha=0.8)
  plt.xlabel('Age')
  plt.ylabel('Charges')
  plt.legend(['Estimate','Actual'])
```

#Let's do that manually

```python
import numpy as np
```

```python
def rmse(targets,predictions):
  return np.sqrt(np.mean(np.square(targets-predictions)))
```

```python
rmse(targets,predictions)
```

```python
w=50
b=100
try_parameters(w,b)
```

```python
targets = non_smoker_df['charges']
predicted = estimate_charges(non_smoker_df.age,w,b)
```

#So,the prediction actually differes from the catual target by  $8461

```python
def try_parameters(w,b):
  ages = non_smoker_df.age
  target = non_smoker_df.charges
  predictions = estimate_charges(ages,w,b)

  plt.plot(ages,predictions,'r',alpha=0.9)
  plt.scatter(ages,target,alpha=0.8,s=8)
  plt.xlabel('Age')
  plt.ylabel('Charges')
  plt.legend(['Predictions','Actual']);

  loss_value = rmse(target,predictions)
  print('RMSE Loss',loss_value)
```

```python
try_parameters(50,100)
```

```python
try_parameters(200,270)
```
